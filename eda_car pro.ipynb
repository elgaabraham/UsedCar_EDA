# Pairplot

```python
#LabelEncoder can be used to normalize labels. It can also be used to transform non-numerical labels to numerical labels.

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
labelencoder= LabelEncoder()
```

```python
df['Year'] = labelencoder.fit_transform(df['Year'])
df['Location'] = labelencoder.fit_transform(df['Location'])
df['Fuel_Type'] = labelencoder.fit_transform(df['Fuel_Type'])
df['Mileage'] = labelencoder.fit_transform(df['Mileage'])
```

```python
new = df.loc[(df['Location'].astype(str)) == '0']
```

```python
sns.set()
columns = ['Name','Location','Year','','Engine']
data_crop = pd.DataFrame(df, columns=columns)
sns.pairplot(data_crop [columns], height = 3 ,kind ='scatter',diag_kind='kde')
plt.show()
```
